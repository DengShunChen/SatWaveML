```python
import matplotlib.pyplot as plt
#import cartopy.crs as ccrs
import seaborn as sns
from buoy_data import Buoy_Data
import numpy as np
```

```python
bd = Buoy_Data('../../')
df = bd.get_data('buoy_data.sqlite.p8.db')
```

```python
pdf = df[(~np.isnan(df['Vm'])) & (~np.isnan(df['H'])) & (~np.isnan(df['Vmsat'])) ]
pdf
```

```python
stnlist=['新竹浮標','龜山島浮標','臺東外洋浮標','龍洞浮標','小琉球浮標','花蓮浮標','東沙島浮標','馬祖浮標']
adf = pdf[pdf['STN'].isin(stnlist)]
adf
```

```python
plt.figure(figsize=(32,32),dpi=200)

sns.set_style("whitegrid")
sns.set(font=['sans-serif'])
sns.set_style("whitegrid",{"font.sans-serif":['Taipei Sans TC Beta']})
sns.lmplot(data=adf[adf['STN']=='東沙島浮標'], x="Vmsat", y="Vm", col="STN", hue="STN", col_wrap=3, height=3, palette="husl",order=1,line_kws={'color':'black'})
plt.tight_layout()
plt.ylim(0,30)
plt.xlim(0,30)
plt.savefig('VmsatVm_scatter.png',dpi=200)
```

```python
def annotate(data,order=1,x='Vm',y='H', **kws):
    ax = plt.gca()
    fit = np.polyfit(data[x], data[y],order)
    if len(fit) == 2:
        a = fit[0]
        b = fit[1]
        ax.text(.05, .8, r'$ %s = %5.2f*%s + %5.2f $' % (y,a,x,b),transform=ax.transAxes)
    elif len(fit) == 3:
        a = fit[0]
        b = fit[1]
        c = fit[2]
        ax.text(.05, .8, r'$ %s = %5.2f*%s^2 + %5.2f*%s + %5.2f $ ' %  (y,a,x,b,x,c),transform=ax.transAxes)
```

```python
sns.lmplot(data=adf, x="Vmsat", y="Vm", palette="husl",order=2,line_kws={'color':'black'})
annotate(adf,2,x="Vmsat",y="Vm")
plt.tight_layout()
plt.savefig('Vmsat_Vm_scatter_all.png',dpi=200)
```

```python
sns.lmplot(data=adf, x="Vmsat", y="Vm", palette="husl",order=1,line_kws={'color':'black'})
annotate(adf,1,x="Vmsat",y="Vm")
plt.tight_layout()
plt.savefig('Vmsat_Vm_scatter_all_1order.png',dpi=200)
```

```python
sns.lmplot(data=adf, x="Vmsat", y="H", palette="husl",order=2,line_kws={'color':'black'})
annotate(adf,2,x="Vmsat",y="H")
plt.tight_layout()
plt.savefig('Vmsat_H_scatter_all.png',dpi=200)
```
